```python
df.drop(columns=['Time on Website','Avatar'],axis=1,inplace=True)
```

```python
X=df.drop(columns='Yearly Amount Spent',axis=1)
Y=df['Yearly Amount Spent']
```

**8) TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=105)
```

**9) TRAINING MODEL**

**KNN REGRESSION**

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Train KNN regression model
k = 7
knn = KNeighborsRegressor(n_neighbors=k)
knn.fit(X_train, Y_train)

# Predict on test set
y_pred = knn.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, y_pred)

# Calculate R-squared error
r2 = r2_score(Y_test, y_pred)

print(f"KNN regression with k={k}")
print(f"Mean squared error: {mse:.2f}")
print(f"R-squared error: {r2:.2f}")
```

**10)MODEL EVALUATION**

```python
from sklearn.metrics import mean_squared_error, r2_score

# Assuming you have true labels (y_true) and predicted labels (y_pred)
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print(f"Mean squared error: {mse:.2f}")
print(f"R-squared score: {r2:.2f}")
```

**DECISION TREE REGRESSOR**

**MODEL EVALUATION**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Train Decision Tree Regression model
max_depth = 11
dtr = DecisionTreeRegressor(max_depth=max_depth)
dtr.fit(X_train, Y_train)

# Predict on test set
y_pred = dtr.predict(X_test)

# Calculate mean absolute error
mae = mean_absolute_error(Y_test, y_pred)

# Calculate mean squared error
mse = mean_squared_error(Y_test, y_pred)

# Calculate R-squared score
r2 = r2_score(Y_test, y_pred)

print(f"Decision Tree Regression with max depth={max_depth}")
print(f"Mean absolute error: {mae:.2f}")
print(f"Mean squared error: {mse:.2f}")
print(f"R-squared score: {r2:.2f}")
```

**RIDGE REGRESSION**

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Train Ridge Regression model
alpha = 0.1
ridge = Ridge(alpha=alpha)
ridge.fit(X_train, Y_train)

# Predict on test set
y_pred = ridge.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, y_pred)

# Calculate mean absolute error
mae = mean_absolute_error(Y_test, y_pred)

# Calculate R-squared score
r2 = r2_score(Y_test, y_pred)

print(f"Ridge Regression with alpha={alpha}")
print(f"Mean squared error: {mse:.2f}")
print(f"Mean absolute error: {mae:.2f}")
print(f"R-squared score: {r2:.2f}")
```

**LASSO REGRESSION**

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Train Lasso Regression model
alpha = 0.1
lasso = Lasso(alpha=alpha)
lasso.fit(X_train, Y_train)

# Predict on test set
y_pred = lasso.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, y_pred)

# Calculate mean absolute error
mae = mean_absolute_error(Y_test, y_pred)

# Calculate R-squared score
r2 = r2_score(Y_test, y_pred)

print(f"Lasso Regression with alpha={alpha}")
print(f"Mean squared error: {mse:.2f}")
print(f"Mean absolute error: {mae:.2f}")
print(f"R-squared score: {r2:.2f}")
```
