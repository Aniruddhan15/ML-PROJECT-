STEP 5: DATA PREPROCESSING

```python
df.shape
```

```python
#1 unique way of GENDER COLUMN
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
#HANDLE CATEGORICAL FEATURE AGE
df['Age'].unique()
```

```python
#Now Encoding
#1way : pd.get_dummies(df['Age'],drop_first=True)

#Targetting ordinal encoding (in sorting way in a consecutive way)
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head(13)
```

```python
#for the test data we need to do transform
```

```python
#Now we ocnvert City_Category
city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
#now we combine the above dataframe
df=pd.concat([df,city],axis=1)
df.head()
```
