TREATMENT OF OUTLIERS

1.   TRIMMING: Remove all outliers
2.   CAPPING: remove apart from particular range


FINDIN OUTLIERS


1.   Using Z score : (CLUE When distplot is bell shaped)
2.   IQR: (CLUE When distplot is skewed)
3. %ile : (CLUE When any other distplot variation occur)



Method 1: Z score

Normal Distribution:
1) within 1 sigma ditribution we have 68% data(34.1+34.1)
2) within 2 sigma distribution we have 95% of data(47.7+47.7)
3) within 3 sigma distribution we have 99.7% of data

**OUTLIERS: Values otuside 3 std deviation(ie: outside of [mu-3(sigma),mu+3(sigma)]**

Refer Notes for Z score
**IQR METHOD**

IQR: Q3-Q1
values < q1-1.5*IQR or values > q3+1.5*IQR are outliers

Maximum limit is Q3
Minimum limit is Q1
