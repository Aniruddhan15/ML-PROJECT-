```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='cp1252')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Amount'].dtypes
```

```python
df['Amount']=df['Amount'].fillna(df['Amount'].median())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.rename(columns={'Product_Category':'PC','Marital_Status':'MStatus'},inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'Product_Category':'PC','Marital_Status':'MStatus','Product_ID':'PId','User_ID':'UId'},inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['MStatus']=df['MStatus'].astype('str')
```

```python
df.describe()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```
